Next, we`ll check for the <b>Price distribution</b> with the help of Histogram

```python
# Define the datasets
datasets = {"AAPL": aapl_data, "AMZN": amzn_data, "GOOGL": googl_data, "IBM": ibm_data, "META": meta_data, 
            "MSFT": msft_data, "NFLX": nflx_data, "NVDA": nvda_data, "ORCL": orcl_data, "TSLA": tsla_data}

# Create a grid layout (2 rows, 5 columns)
fig, axes = plt.subplots(2, 5, figsize=(18, 10))
fig.suptitle("Stock Price Distribution from various markets", fontsize=16)

# Iterate through datasets and plot histograms
for ax, (name, data) in zip(axes.flat, datasets.items()):
    sns.histplot(data["Close"], kde=True, bins=30, color="blue", ax=ax)
    ax.set_title(name)
    ax.set_xlabel("Stock Price")
    ax.set_ylabel("Frequency")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

As we can see, growth stocks,which are Tesla, Nvidia, META, Netflix, Amazon, Microsoft and Google  are → Right-skewed distributions which indicate that these stocks had lower values historically but grew significantly.
Stable Stocks like the IBM and Oracle are → More symmetric or bell-shaped distributions which suggest that stock prices have fluctuated but remained within a range.

It is notably seeable that all this companies in <b>recent </b>years have had a massive increase in stock prices

There is notably a peak at a lower stock price, suggesting that Apple's stock price has grown over time, but historical data includes many lower values.
We can also tell by this that the stock prices are right-skewed, meaning lower prices were more frequent.



Moreover, we can cover<b> Volatility Analysis,</b> to analyze daily percentage price changes to assess volatility.

```python
# Plot individual volatility graphs
for name, data in datasets.items():
    data["Daily_Return"] = data["Close"].pct_change()  # Calculate daily returns (percentage change)
    
    plt.figure(figsize=(10, 5))
    plt.plot(data["Date"], data["Daily_Return"], label=f"{name} Volatility", color="red")
    plt.axhline(0, linestyle="--", color="black", linewidth=1)  # Add a zero line for reference
    plt.title(f"Daily Return Volatility: {name}")
    plt.xlabel("Date")
    plt.ylabel("Daily Return (%)")
    plt.legend()
    plt.show()
```

As we can see from the graphs above, that companies that have been on incrimental rise over the recent years have significantly shown high volatility.
Higher volatility means higher risk but also higher potential returns.
Examples are the Apple stock market,Microsoft,Tesla anf Oracle

This analysis helps in building predictive models by identifying high-volatility periods where price swings are significant.

Lastly, we can <b>partake Correlation Analysis</b> which will help investigate correlations between the stocks to understand their relationships.

```python
# Combine closing prices for correlation analysis
combined_data = pd.concat([aapl_data.set_index("Date")["Close"],
                          amzn_data.set_index("Date")["Close"],
                          googl_data.set_index("Date")["Close"],
                          ibm_data.set_index("Date")["Close"],
                          meta_data.set_index("Date")["Close"],
                          msft_data.set_index("Date")["Close"],
                          nflx_data.set_index("Date")["Close"],
                          nvda_data.set_index("Date")["Close"],
                          orcl_data.set_index("Date")["Close"],
                          tsla_data.set_index("Date")["Close"]], axis=1)

combined_data.columns = ["AAPL", "AMZN", "GOOGL", "IBM", "META", "MSFT", "NFLX", "NVDA", "ORCL", "TSLA"]

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(combined_data.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Stock Prices")
plt.show()
```

From this heatmap, we can visually understand that,
1. The Red shades (closer to +1) indicate strong positive correlations as this tells us that the stocks move together.
2. The Blue shades (closer to -1) indicate strong negative correlations and therefore tell us the stocks move in negative diection
3. The White/light colors (close to 0) indicate weak or no correlation.



Highly correlated stocks, like the (Apple- Microsoft) and the (Amazon - Google) indicate that they move together,
<b>this means that when the price of one stock goes up or down, the other stock tends to increase or decrease in a similar pattern</b>. They indicate a high positive correlation (close to +1).

IBM shows negative correlation with most tech stocks (-0.50 to -0.74). This suggests IBM behaves differently from most other stocks in this selection.

Most tech stocks, these are Apple, Amazon, Google, Micrososft, Nvidia, Oracle and Tesla show strong positive correlations (above 0.9), meaning they likely follow similar trends.
This is notable as we have previously noted that this companies tend to have gotten an incremental rise in stock prices in recent years.
